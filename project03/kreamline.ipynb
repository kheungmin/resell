```python
def price_preprocessing(data):
    data = data.replace('-', '0')
    data = data.replace(',', '')
    
    return int(data)
```

```python
def date_preprocessing(data):
    data = data.strip()
    if data != '-':
        if int(data[:2]) < 50:
            data = '20' + data
        elif int(data[:2]) >= 50:
            data = '19' + data
    
    data = data.replace('-', '1900/01/01')

    return data
```

```python
def release_price_preprocessing(data):
    filtered = re.findall('\d[0-9,]*원', data)
    if filtered:
        data = filtered[0].replace('원', '').replace(',', '')
    else:
        data = '0'
    
    return int(data)
```

```python
df_copy['rescent_price'] = df_copy['rescent_price'].apply(price_preprocessing)
df_copy['instant_buy_price'] = df_copy['instant_buy_price'].apply(price_preprocessing)
df_copy['instant_sell_price'] = df_copy['instant_sell_price'].apply(price_preprocessing)
df_copy['release_date'] = df_copy['release_date'].apply(date_preprocessing)
df_copy['release_price'] = df_copy['release_price'].apply(release_price_preprocessing)
```
