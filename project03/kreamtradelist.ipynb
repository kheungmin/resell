```python
user = 'team04'
pw = '1111'
host = 'ec2-54-95-8-243.ap-northeast-1.compute.amazonaws.com'
client = pymongo.MongoClient(f'mongodb://{user}:{pw}@{host}:27017/')
db = client.resell
```

```python
dft = pd.DataFrame(db.trade.find())
dft
```

```python
dfk = pd.DataFrame(db.kream.find())
dfk.head()
```

```python
def change_num(data):
    data = data[3:]
    if data == '':
        data = '0'
    if '.' in data:
        data = data.replace('만', '000')
    else:
        data = data.replace('만', '0000')
    data = data.replace('.', '').replace(',', '')
    data = int(data)
    
    return data
```

```python
dfk['all_trade'] = dfk['all_trade'].apply(change_num)
```
