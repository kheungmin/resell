```python
import pymongo
import pandas as pd
```

```python
user = 'team04'
pw = '1111'
host = 'ec2-54-95-8-243.ap-northeast-1.compute.amazonaws.com'
client = pymongo.MongoClient(f'mongodb://{user}:{pw}@{host}:27017/')
db = client.resell
```

```python
df = pd.DataFrame(db.kream.find())
df
```

```python
df2 = df
df2
```

```python
def change_num(data):
    data = data[3:]
    if data == '':
        data = '0'
    if '.' in data:
        data = data.replace('만', '000')
    else:
        data = data.replace('만', '0000')
    data = data.replace('.', '').replace(',', '')
    data = int(data)
    
    return data
```

```python
def change_num2(data):
    if '.' in data:
        data = data.replace('만', '000')
    else:
        data = data.replace('만', '0000')
    data = data.replace('.', '').replace(',', '')
    data = int(data)
    
    return data
```

```python
def change_price(data):
    data = data.replace('-', '0')
    data = int(data.replace('원', '').replace(',', ''))
    return data
```

```python
df2 = df2.fillna('0')
```

```python
df2['price'] = df2['price'].apply(change_price)
df2['like'] = df2['like'].apply(change_num2)
df2['post'] = df2['post'].apply(change_num2)
```

```python
df2['all_trade'] = df2['all_trade'].apply(change_num)
```
